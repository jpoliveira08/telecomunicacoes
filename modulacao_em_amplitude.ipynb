<h2><li><bold>Introdução</bold></li></h2>
<p>
A modulação em amplitude (AM) desempenha um papel crucial na transmissão de sinais de áudio e dados em diferentes sistemas de comunicação, sendo uma técnica fundamental na modulação de sinais. O presente trabalho tem como objetivo modular um sinal de áudio em relação a uma portadora, utilizando os conceitos de modulação em amplitude.
</p>

<h2><li><bold>Problema</bold></li></h2>

Considerando um sinal de áudio $15sin(2\pi 1500t)$ e o sinal da portadora sendo definido por $60sin(2\pi 100000t)$. Será realizado o esboço dos sinais de áudio, portadora, assim como da onda modulada. Para desenvolver os conceitos associado ao tópico de modulação em amplitude, será determinado o fator e percentual de modulação, identificação das frequências dos sinais de áudio e portadora e por fim determinado as frequências presentes no espectro de Fourier da onda modulada.
</ol>


```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format='retina'
plt.rcParams['figure.figsize'] = (16, 6)
```

```python
angularFrequency = lambda frequency: (2 * np.pi * frequency)

signalAmplitude = 15;
carrierAmplitude = 60;
time = np.arange(0, 0.002, 1e-7)
```

```python
audioSignal = signalAmplitude*np.sin(angularFrequency(1500)*time)
plt.plot(time, audioSignal)
plt.show()
```

```python
carrierSignal = carrierAmplitude*np.sin(angularFrequency(100000)*time)
plt.plot(time, carrierSignal)
plt.show()
```

```python
modulatedSignal = carrierSignal * (1 + (audioSignal/carrierAmplitude))
plt.plot(time, modulatedSignal)
maxMoludatedSignal = max(modulatedSignal)
plt.axhline(maxMoludatedSignal, color='r', linestyle='--', label=f'Máximo: {maxMoludatedSignal:.2f}')
plt.show()
```

```python
# Sinal de áudio
plt.subplot(3, 1, 1)
plt.plot(time, audioSignal)
plt.title('Sinal de áudio')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Sinal da portadora
plt.subplot(3, 1, 2)
plt.subplots_adjust(hspace=0.7)
plt.plot(time, carrierSignal)
plt.title('Sinal da portadora')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Sinal modulado
plt.subplot(3, 1, 3)
plt.subplots_adjust(hspace=0.7)
plt.plot(time, modulatedSignal)
maxMoludatedSignal = max(modulatedSignal)
plt.axhline(maxMoludatedSignal, color='r', linestyle='--', label=f'{maxMoludatedSignal:.2f}')
plt.title('Sinal modulado')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.legend(fontsize=6)
plt.grid(True)
```

Podemos encontrar o fator de modulação e o percentual de modulação por meio das equaçãoes apresentadas a seguir.
$$µ=Am/Ac$$

$$µ100=µ*100%$$


```python
fatordemod = (signalAmplitude)/(carrierAmplitude);
percmod = fatordemod * 100%
```
