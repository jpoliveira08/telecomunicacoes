<h2><li><bold>Introdução</bold></li></h2>
<p>
A modulação em amplitude (AM) desempenha um papel essencial na transmissão de sinais de áudio e dados em diversos sistemas de comunicação, sendo uma técnica fundamental na modulação de sinais. Este estudo tem como propósito a modulação de um sinal de áudio em relação a uma portadora, empregando os princípios da modulação em amplitude.

Já a demodulação em amplitude (AM) é um processo vital em sistemas de comunicação que utilizam a modulação em amplitude para transmitir sinais de áudio e dados. Essa etapa tem a função inversa da modulação, sendo responsável por extrair o sinal original após a transmissão. Durante a demodulação AM, o sinal modulado em amplitude é recebido e, em seguida, é realizado o processo de recuperação do sinal original. Diversas técnicas podem ser empregadas para realizar essa demodulação, como o uso de detectores de envoltória ou a utilização de circuitos específicos. A precisão e eficiência desse processo são cruciais para garantir uma recuperação precisa do sinal original, contribuindo para a fidelidade na transmissão de informações em sistemas de comunicação que empregam a modulação em amplitude.

<h2><li><bold>Somador</bold></li></h2>
A seção do circuito responsável por aplicar a onda modulante (Vm) à portadora (Vc) é o somador. Esta parte consiste em uma rede resistiva de mixagem, composta por três resistores, sendo eles, R1 responsável pelo sinal modulante, R2 pelo sinal da portadora e R3 sendo o sinal resultante da operação, em que a tensão sobre R3 podendo ser avaliada através das correntes como visto na equação abaixo:

$$
V_{R3} = (I_c + I_m)R_3
$$

Onde Ic e Im podem ser encontradas substituindo respectivamente por
(Vc − VR3)/R1 e (Vm − VR3)/R2. Isolando VR3 se chega em:

$$
V_{R3} = (\frac{V_c}{R_1} + \frac{V_m}{R_2})(\frac{R_1R_2R_3}{R_1+R_2+R_3})
$$

Como R1 e R2 são iguais a equação pode ser simplificada, percebe-se como os sinais são somados nesta fase.

$$
V_{R3} = (\frac{R_1R_3}{2R_1+R_3})(V_c+V_m) = K(V_c+V_m)
$$

Sabendo que a série de Fourier dessa equação é:

$$
V_{R3} = \frac{a_0}{2} + \sum\limits_{n=1}^{∞}a_ncos(nt)+\sum\limits_{n=1}^{∞}b_nsen(nt)
$$

E sabendo que os
coeficientes no domínio do tempo pode ser definidos por
$V_{R3} = V_{R31} + V_{R32}$

Expandindo para o coeficiente $V_{R31}$

$$
V_{R31}(t) = \sum\limits_{n=1}^{∞} \frac{47,87}{2\pi} cos(nt) [-2\pi cos(2\pi(200*10^{3}+n))-2\pi cos(2\pi (200*10^{3}-n))]+\sum\limits_{n=1}^{∞}\frac{47,87}{\pi} sen(nt)[(\frac{2\pi sen(200*10^{3} - n)}{\pi 400*10^{3}-2n})-(\frac{2\pi sen(200*10^{3} + n)}{\pi 400*10^{3}+2n})]
$$

<h2><li><bold>Retificador</bold></li></h2>

O diodo desempenha um papel crucial ao eliminar a parte negativa do sinal e permitir a multiplicação dos sinais. Essa função é essencial para a geração de banda dupla, conforme será demonstrado na aplicação do circuito, além de facilitar a integração do sinal modulante na portadora. Nesse contexto, o diodo pode ser modelado como um dispositivo de lei quadrada, conforme indicado na Equação abaixo:

$$
I_D = a_1 V_{R3} + a_2 {V_{R3}}^2
$$

Como a tensão pode ser dada por $I_DR_4$, tem se a Equação

$$
V_{R4} = I_DR_4 = R_4(a_1K_1(V_c + V_m) + a_2{K_1}^2(V_c + V_m)^2)
$$

Substituindo $V_c$ e $V_m$ respectivamente por $v_ccos(ω_ct)$ e $v_mcos(ω_mt)$. Substituindo e
aplicando as identidades trigonométricas tem se a Equação

$$
V_{R4} = b_1(v_ccos(ω_ct) + v_mcos(ω_mt))+ b_2 (v_2c_2 \frac{{V_c}^2}{2} (1 + cos(2ω_ct)) + v_cv_m(cos((ω_c + ω_m)t) + cos((ω_c − ω_m)t)) + \frac{{V_m}^2}{2}(1 + cos(2ω_mt))
$$

São identificadas seis frequências distintas resultantes, juntamente com componentes CC. No entanto, a faixa relevante para o sinal está concentrada principalmente na proximidade de ωc, a frequência da portadora. Os coeficientes da série de Fourier podem ser calculados analisando a Equação anterior. Ao adotar um valor de η=1 para o fator de idealidade construtiva, podemos deduzir as constantes nas equações a seguir e inseri-las nas funções em relação ao tempo.

$$
a_1 =\frac{I_s}{ηVt}=\frac{6, 9 ∗ 10^{−16}}{1 ∗ 26 ∗ 10^{−3}}= 2, 65 ∗ 10^{−14}
$$


$$
a_2 =\frac{I_s}{ηVt}=\frac{6, 9 ∗ 10^{−16}}{2*{(1 ∗ 26 ∗ 10^{−3})}^2}= 5, 1 ∗ 10^{−13}
$$

$$
b_1 = R_4a_1K_1 = 100 ∗ 103∗ 2, 65 ∗ 10^{−14} ∗ 4, 878 = 1, 3 ∗ 10^{−8}
$$
$$
b_1 = R_4a_2K_2 = 100 ∗ 103 ∗ 5, 1 ∗ 10^{−13} ∗ 4, 8782 = 1, 21 ∗ 10^{−6}
$$

<h2><li><bold>Circuito ressonante</bold></li></h2>

Para capturar apenas a faixa de frequência ωc, empregamos um circuito ressonante projetado especificamente para essa frequência. A função do filtro no domínio da frequência é expressa pela seguinte equação:

$$
V_{out}(ω)=\frac{jω}{R_5(1 − {ω}^2CL)jωL}*V_{R4}(ω)
$$

Com essas informações, é possível determinar a frequência que deve ser ajustada. Ao filtrar $V_{R4}$, a saída aproximadamente segue a seguinte equação:

$$
V_{out}(t) = b_1v_ccos(ω_ct) + b_2v_cv_m(cos((ω_c + ω_m)t) + cos((ω_c − ω_m)t))
$$

# Representação do código matemático da modulação e demodulação

A seguir será apresenta o código responsável por realizar a modulação de um sinal AM, assim como sua respectiva demodulação utilizando a representação em código de um circuito detector à diodo.
Na etapa a seguir do código é realizado a definição das variáveis e de uma função que será utilizada nos passos seguintes.


```python
angularFrequency = lambda frequency: (2 * np.pi * frequency)

signalAmplitude = 8;
carrierAmplitude = 10;
time = np.arange(0, 0.002, 2e-7)
```

Em sequência é realizado a montagem do sinal de áudio, assim como a plotagem do seu respectivo formato.

```python
audioSignal = signalAmplitude*np.sin(angularFrequency(5000)*time)
plt.plot(time, audioSignal)
plt.show()
```

E neste ponto, é realizado a montagem do sinal da portadora, assim como sua representação gráfica.

```python
carrierSignal = carrierAmplitude*np.sin(angularFrequency(100000)*time)
plt.plot(time, carrierSignal)
plt.show()
```

Em seguida, será realizado o cálculo e a representação gráfica da onda modulada.

```python
modulatedSignal = carrierSignal * (1 + (audioSignal/carrierAmplitude))
m = signalAmplitude / carrierAmplitude
modulatedSignal = (1 + m * audioSignal / signalAmplitude) * carrierSignal

plt.plot(time, modulatedSignal)
maxMoludatedSignal = max(modulatedSignal)
plt.axhline(maxMoludatedSignal, color='r', linestyle='--', label=f'Máximo: {maxMoludatedSignal:.2f}')
plt.show()
```

```python
# Sinal de áudio
plt.subplot(3, 1, 1)
plt.plot(time, audioSignal)
plt.title('Sinal de áudio')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Sinal da portadora
plt.subplot(3, 1, 2)
plt.subplots_adjust(hspace=0.7)
plt.plot(time, carrierSignal)
plt.title('Sinal da portadora')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Sinal modulado
plt.subplot(3, 1, 3)
plt.subplots_adjust(hspace=0.7)
plt.plot(time, modulatedSignal)
maxMoludatedSignal = max(modulatedSignal)
plt.axhline(maxMoludatedSignal, color='r', linestyle='--', label=f'{maxMoludatedSignal:.2f}')
plt.title('Sinal modulado')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.legend(fontsize=6)
plt.grid(True)
```

```python
# Frequências de Fourier

fourierTransform = np.fft.fft(modulatedSignal)
fourierFrequency = np.fft.fftfreq(len(modulatedSignal), 2e-7)

plt.subplot(2, 1, 1)
plt.plot(time, modulatedSignal)
plt.title('Sinal Modulado')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(fourierFrequency, np.abs(fourierTransform))
plt.title('Frequência de Fourier')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.xlim(-200000, 200000)
plt.show()
```

Em sequência será implementado o demodulador com diodo detector de envelope. O detector de envelope é composto por um diodo que retira a metade positiva do sinal recebido, anulando a outra metade. Em seguida, um capacitor é utilizado para remover a componente de alta frequência, atuando como um filtro passa-baixa. Em seguida, será apresentado o mesmo resultado da figura 6, e pode-se concluir que matemáticamente não há o pequeno offset presente na figura 6, que impede o sinal de tocar no eixo horizontal.

```python
from scipy.signal import butter,filtfilt

modulatedSignalToDemodulate = modulatedSignal

modulatedSignalToDemodulate[modulatedSignalToDemodulate < 0] = 0

plt.plot(time, modulatedSignalToDemodulate)
plt.show()
```

```python
# Implementação do filtro
import scipy.fftpack as sf
import scipy.signal as sig

R = 2.7e3
C = 1e-9
frequenciaAmostragem = 2e6

frequenciaCorte = 1 / (2 * np.pi * R * C)

w = frequenciaCorte / (frequenciaAmostragem / 2) # normalizando a frequencia
b, a = sig.butter(1, w, 'low')
sinalFiltrado = sig.filtfilt(b, a, modulatedSignalToDemodulate)

W, h = sig.freqz(b, a, worN = 1024)
W = frequenciaAmostragem * W / (2 * np.pi)

plt.subplot(2, 1, 1)
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Resposta em frequência do filtro')
plt.plot(W, 20 * np.log10(h))

sinalFiltrado *= 2 / signalAmplitude

plt.subplot(2, 1, 2)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal Filtrado')
plt.plot(time, sinalFiltrado)

plt.tight_layout()
plt.show()
```
